```python

import matplotlib.pyplot as plt
import numpy as np
from ecgdetectors import Detectors

from EcdDataConverter import import_mat

data = import_mat('./files/JS00001.hea', './files/JS00001.mat')
II = data.II
length = len(II)
t = 1 / data.frequency
time = np.arange(t, length * t + t, t)
plt.plot(time, II)
plt.xlim(0, 3)
plt.show()
```

```python
detector = Detectors(data.frequency)
result = detector.wqrs_detector(II)
result = [r * t for r in result]
plt.vlines(x=result, ymin=min(II), ymax=max(II), color="r")
plt.plot(time, II)
plt.xlim(0, 3)
plt.show()
```
