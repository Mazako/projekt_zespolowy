```python
def plot_detector_results(ax, detector_results, title, time, canal):
    ax.vlines(x=detector_results, ymin=min(canal), ymax=max(canal), color="r")
    ax.plot(time, canal)
    ax.set_xlim(0, 3)
    ax.set_title(title)
    
import matplotlib.pyplot as plt

from EcdDataConverter import import_mat
from QrsDetector import *
data = import_mat('./files/JS00005.hea', './files/JS00005.mat')
channel = nk.ecg_clean(data.I, data.frequency)

wqrs = find_wqrs(channel, data.frequency)
neurokit = find_neurokit(channel, data.frequency)
kalidas = find_kalidas(channel, data.frequency)



fig, ax = plt.subplots(3, 1, figsize=(15, 10))

plot_detector_results(ax[0], wqrs, 'WQRS', data.time, channel)
plot_detector_results(ax[1], neurokit, 'neurokit', data.time, channel)
plot_detector_results(ax[2], kalidas, 'kalidas', data.time, channel)
```
