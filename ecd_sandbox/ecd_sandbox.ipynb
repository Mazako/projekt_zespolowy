```python
import scipy
import numpy as np
import matplotlib.pyplot as plt 
from QRS_detector import detect
data = scipy.io.loadmat('./files/JS00001.mat')

II = data['val'][1]
II = np.array([x / 1000 for x in II])
length = len(II)
f = 500 #Hz
t = 1 / f
time = np.arange(t, length * t + t, t)
plt.plot(time, II)
plt.xlim(0, 3)
plt.show()
```

```python
ecd = detect(II, f)
ecd_time = [peak * t for peak in ecd]
plt.vlines(x=ecd_time, ymin=min(II), ymax=max(II), color="r")
plt.plot(time, II)
plt.xlim(0, 3)
plt.show()
```
