```python
import neurokit2 as nk
import EcdDataConverter
import matplotlib.pyplot as plt

data = EcdDataConverter.import_mat('./files/JS00005.hea ', './files/JS00005.mat')

channel = data.I
time = data.time

frame = nk.ecg_process(channel, data.frequency)[0]

cleaned_signal = frame['ECG_Clean']
R_Peaks = frame[frame['ECG_R_Peaks'] == 1].index.tolist()
P_Peaks = frame[frame['ECG_P_Peaks'] == 1].index.tolist()
Q_Peaks = frame[frame['ECG_Q_Peaks'] == 1].index.tolist()
S_Peaks = frame[frame['ECG_S_Peaks'] == 1].index.tolist()
T_Peaks = frame[frame['ECG_T_Peaks'] == 1].index.tolist()

plt.plot(cleaned_signal)
plt.vlines(x=R_Peaks, ymin=min(cleaned_signal), ymax=max(cleaned_signal), color="r")
plt.vlines(x=P_Peaks, ymin=min(cleaned_signal), ymax=max(cleaned_signal), color="g")
plt.vlines(x=Q_Peaks, ymin=min(cleaned_signal), ymax=max(cleaned_signal), color="b")
plt.vlines(x=S_Peaks, ymin=min(cleaned_signal), ymax=max(cleaned_signal), color="y")
plt.vlines(x=T_Peaks, ymin=min(cleaned_signal), ymax=max(cleaned_signal), color="k")
plt.xlim(100, 400)
plt.show()
```

```python
bpm = frame['ECG_Rate'].mean()
print(f'Przeciętne BPM: {bpm}')
```
