```python
import neurokit2 as nk
import EcdDataConverter
import matplotlib.pyplot as plt

data = EcdDataConverter.import_mat('./files/JS00001.hea ', './files/JS00001.mat')

channel = data.I
time = data.time

frame = nk.ecg_process(channel, data.frequency)[0]

cleaned_signal = frame['ECG_Clean']

P_Peaks = frame[frame['ECG_R_Peaks'] == 1].index.tolist()
# Q_Peaks = frame[frame['ECG_Q_Peaks'] == 1].index.tolist()
P_onsets = list(frame[frame['ECG_R_Onsets'] == 1].index)
P_offsets = list(frame[frame['ECG_R_Offsets'] == 1].index)

print(len(P_Peaks), len(P_onsets), len(P_offsets))

# S_Peaks = frame[frame['ECG_S_Peaks'] == 1].index.tolist()
# T_Peaks = frame[frame['ECG_T_Peaks'] == 1].index.tolist()

plt.plot(cleaned_signal)
plt.vlines(x=P_Peaks, ymin=min(cleaned_signal), ymax=max(cleaned_signal), color="r")
plt.vlines(x=P_onsets, ymin=min(cleaned_signal), ymax=max(cleaned_signal), color="g")
plt.vlines(x=P_offsets, ymin=min(cleaned_signal), ymax=max(cleaned_signal), color="b")
# # plt.vlines(x=S_Peaks, ymin=min(cleaned_signal), ymax=max(cleaned_signal), color="y")
# # plt.vlines(x=T_Peaks, ymin=min(cleaned_signal), ymax=max(cleaned_signal), color="k")
plt.xlim(1000, 2000)
# plt.show()
```
